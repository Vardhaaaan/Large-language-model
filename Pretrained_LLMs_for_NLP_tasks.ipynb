```python
from transformers import pipeline

# Load pre-trained sentiment classification model
classifier = pipeline(
    "text-classification",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Paragraph input
paragraph = """
The movie had a strong storyline and excellent performances.
However, the pacing was slow in some parts, but overall it was enjoyable.
"""

# Perform classification
output = classifier(paragraph)

# Display result
print("Classification Result:", output)
```

```python
from transformers import pipeline

# Load pre-trained summarization model
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

# Paragraph input
paragraph = """
Artificial Intelligence is transforming industries by enabling machines to learn from data,
make decisions, and perform tasks that typically require human intelligence. It is widely used
in healthcare, finance, education, and autonomous systems. However, challenges such as data
privacy, bias, and ethical concerns must be addressed to ensure responsible AI development.
"""

# Generate summary
summary = summarizer(paragraph, max_length=60, min_length=30, do_sample=False)

# Print summary
print("Summary:", summary[0]["summary_text"])
```

```python
from transformers import pipeline

# Load pre-trained NER model
ner = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

# Paragraph input
paragraph = """
Sundar Pichai is the CEO of Google and Alphabet Inc.
He was born in India and currently lives in the United States.
"""

# Perform Named Entity Recognition
entities = ner(paragraph)

# Print extracted entities
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}")
```

```python
from transformers import pipeline

# Load pre-trained Question Answering model
qa_model = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

# Context paragraph
context = """
Artificial Intelligence enables machines to learn from data and perform tasks
that typically require human intelligence. AI is widely used in healthcare,
finance, education, and autonomous systems.
"""

# Take question from user
question = input("Enter your question: ")

# Get answer
result = qa_model(
    question=question,
    context=context
)

# Print answer
print("Answer:", result["answer"])
```
