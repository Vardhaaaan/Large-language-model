```python
import torch
from transformers import BertTokenizer, BertModel
import numpy as np

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

sentence = "Every moment is beginning"

inputs = tokenizer(sentence, return_tensors="pt")

with torch.no_grad():
    outputs = model(**inputs)

# Full token embeddings
embeddings = outputs.last_hidden_state.squeeze(0)  # shape: (5, 768)

tokens = tokenizer.tokenize(sentence)

for i, token in enumerate(tokens):
    print(f"\nToken: {token}")
    print(embeddings[i].numpy())   # FULL 768-D vector
```

```python
import torch
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

sentence = "Every moment is beginning"

inputs = tokenizer(sentence, return_tensors="pt")

# Get position IDs
position_ids = torch.arange(inputs["input_ids"].size(1)).unsqueeze(0)

# Extract positional embedding layer
with torch.no_grad():
    position_embeddings = model.embeddings.position_embeddings(position_ids)

tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

print("Positional Embedding Shape:", position_embeddings.shape)

for i, token in enumerate(tokens):
    print(f"\nToken: {token}")
    print(position_embeddings[0][i].numpy())  # FULL 768 values
```
